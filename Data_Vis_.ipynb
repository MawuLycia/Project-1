```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('breast_cancer_modified.csv')
```

```python
data.head()
```

```python
col =data.columns
print(col)
```

```python
y =data.Diagnosis
list = ['ID','Diagnosis']
x =data.drop(list, axis =1)
x.head()
```

```python
ax =sns.countplot(y, label ="Count")
B, M =y.value_counts()
print('Number of Benign:',B)
print('Number of Malignant :', M)
```

```python
y = data.Diagnosis.replace({'M': 1, 'B': 0})
```

```python
x.describe()
```

```python
#Creating swarm plots to see variations in each feature better. This is for the breast cancer dataset (Features 1-15)

sns.set(style = "whitegrid", palette = "muted")
data_diag = y
data = x
data_stan = (data - data.mean())/ data.std()      #Standatdization
data = pd.concat([y, data_stan.iloc[ :,0:15]], axis =1 )
data = pd.melt(data, id_vars= "Diagnosis", var_name= "Features", value_name= "value")

plt.figure(figsize = (10,10))
sns.swarmplot(x ="Features", y = "value" ,hue = "Diagnosis", data = data , size = 2 )
plt.xticks(rotation =90)
```

```python
#Creating swarm plots to see variations in each feature better. This is for the breast cancer dataset (Features 16-30)

sns.set(style = "whitegrid", palette = "muted")
data_diag = y
data = x
data_stan = (data - data.mean())/ data.std()      #Standatdization
data = pd.concat([y, data_stan.iloc[ :,16:30]], axis =1 )
data = pd.melt(data, id_vars= "Diagnosis", var_name= "Features", value_name= "value")

plt.figure(figsize = (10,10))
sns.swarmplot(x ="Features", y = "value" ,hue = "Diagnosis", data = data , size = 2 )
plt.xticks(rotation =90)
```

Looking at the swarm plots, Features 8, 21, and 23 have the classes mostly separated and Features 10, 15 and 19 have the classes malignant and benign mixed. will be difficult to classify based on that feature
